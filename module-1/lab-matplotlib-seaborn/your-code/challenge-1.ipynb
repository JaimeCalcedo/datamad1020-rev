**0. Load seaborn dataset "tips"**

```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
tips.sample(10)
```

**1. What is the prefered day for eating?**

```python
sns.histplot(x = tips.day)
#We can see Saturday is the prefered day for eating
```

**2. What days of the week has more men than women payers?**

```python
sns.histplot(x = tips.day, hue = tips.sex)
#Saturdays and Sundays have more men than women payers
```

**3. Do you think bill amounts change considerably depending on smoking?**

```python
sns.kdeplot(x = tips.total_bill, hue = tips.smoker)
```

**4. Is tip correlated to total_bill? Draw a scatter plot**

```python
sns.scatterplot(x = tips.total_bill, y = tips.tip)
#It is correlated, but not as much with big bills. 
```
